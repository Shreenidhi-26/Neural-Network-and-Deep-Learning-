#**LAB-05 : Illustrate the step-by-step process of applying convolutional operations on data for enhanced understanding**

#**Aim**
To illustrate the application of convolutional operations on a 2D matrix using a kernel, and to demonstrate how these operations aid in feature extraction and transformation of data through visualization of the input, kernel, and resulting matrix.



#**Libraries**

* **NumPy:** Handles numerical operations and arrays; used for matrix manipulation.
* **Matplotlib:** Creates visualizations; used to plot matrices and results.
* **OpenCV (cv2):** For image loading and processing.

####Step-by-Step Process

1. Importing Libraries:


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

2. Define the Input Data:

```python
# Option to add an image path
image_path =  '/content/tom-and-jerry.jpg'
if image_path:
    # Load the image in grayscale mode
    input_data = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if input_data is None:
        raise ValueError("Image not found or unable to load.")
    plt.title('Input Image')
else:
    # Define a 2D input matrix (e.g., an image)
    input_data = np.array([
        [0, 1, 2, 3, 0],
        [4, 5, 6, 7, 0],
        [8, 9, 10, 11, 0],
        [0, 1, 2, 3, 0]
    ])
    plt.title('Input Data')

print("Input Data:")
print(input_data)

# Display the input data or image
plt.imshow(input_data, cmap='gray')
plt.colorbar()
plt.show()
```

**Inference:** The process loads an image in grayscale from a specified path or uses a predefined 2D matrix as input, and then displays this data for visualization.

3. Define the Convolutional Kernel (Filter):Define the Convolutional Kernel (Filter):

```python
# Define a simple 2x2 kernel (filter)
kernel = np.array([
    [1, 0],
    [0, -1]
])

print("Kernel:")
print(kernel)

# Display the kernel
plt.imshow(kernel, cmap='gray')
plt.title('Convolutional Kernel')
plt.colorbar()
plt.show()
```

**Inferene:** The code defines a 2x2 convolutional kernel with values `[1, 0]` and `[0, -1]`, then prints and visualizes this kernel using a grayscale colormap. The displayed kernel highlights its structure and values, which are used for feature detection in convolution operations.

4. Perform Convolution Operation:

```python
# Convolution operation: manually applying the kernel over the input matrix
def convolution2d(input_data, kernel):
    input_height, input_width = input_data.shape
    kernel_height, kernel_width = kernel.shape

    output_height = input_height - kernel_height + 1
    output_width = input_width - kernel_width + 1

    output = np.zeros((output_height, output_width))

    for i in range(output_height):
        for j in range(output_width):
            region = input_data[i:i+kernel_height, j:j+kernel_width]
            output[i, j] = np.sum(region * kernel)

    return output

# Apply convolution
conv_result = convolution2d(input_data, kernel)

print("Convolution Result:")
print(conv_result)

# Display the convolved output
plt.imshow(conv_result, cmap='gray')
plt.title('Convolution Result')
plt.colorbar()
plt.show()
```

**Inference:** The code performs a 2D convolution by sliding a kernel over the input matrix, calculating the dot product for each position, and generating an output matrix. The result of this convolution is printed and visualized to show how the kernel modifies the input data.

5. Visualize the Results:

```python
# Plot the original input data, kernel, and convolved result side by side
plt.figure(figsize=(15, 5))

# Original Image or Data
plt.subplot(1, 3, 1)
plt.imshow(input_data, cmap='gray')
plt.title('Input Data' if not image_path else 'Input Image')
plt.axis('off')

# Convolution Kernel
plt.subplot(1, 3, 2)
plt.imshow(kernel, cmap='gray')
plt.title('Convolutional Kernel')
plt.axis('off')

# Convolution Result
plt.subplot(1, 3, 3)
plt.imshow(conv_result, cmap='gray')
plt.title('Convolution Result')
plt.axis('off')

plt.show()
```

**Inference:** The code visualizes the original input data, the convolutional kernel, and the result of the convolution operation side by side. This comparison helps to illustrate how the kernel affects the input data and the resulting feature map.
